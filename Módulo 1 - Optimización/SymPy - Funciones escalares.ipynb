## 1. Teorema de Fermat (análisis)

Si una función $f(x)$ alcanza un máximo o mínimo local en $x=c$, y si la derivada $f'(c)$ existe en el punto $c$, entonces $f'(c) = 0$.

### Ejemplo

Sabemos que la función $f(x)=(x-2)^2$ tiene un mínimo global en $x=2$, pues

$$f(x)=(x-2)^2\geq0,\qquad\text{y}\qquad f(x)=(x-2)^2=0 \qquad\text{si y solo si}\qquad x=2.$$

1. Definir una función simbólica
  1. Importar librerías
  2. Definimos la variable simbólica $x$
  3. Definimos la función $f$
2. Derivar con respecto a $x$. $df=f'(x)$
3. Encontrar la solución para la derivada igualada a cero. Encontramos los puntos críticos $c$, tal que $df(x)=0$
4. Definir si los puntos críticos son máximos o mínimos.

````python
# Paso 1.1
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
````

````python
# Paso 1.2
sym.var('x')
````

````python
# Paso 1.3
f=(x-2)**2  # Operador **  ->  potencia
f           # Mostramos 'f'
````

````python
# Paso 2
df=sym.diff(f,x) # Derivamos 'f' con respecto a 'x'
df
````

````python
# Paso 3
c=sym.solve(df,x)
c   # Esto es un arreglo
````

````python
c[0] # Accedemos al (único) valor
````

Veamos la gráfica...

````python
# Convertir función simbólica f a una función numérica
f_num=sym.lambdify(x,f,'numpy')
````

````python
# Definir coordenas x
x_num=np.linspace(0,4,100) # Definimos el rango de la gráfica alrededor del punto crítico
````

````python
# Graficar:
#    Calcular coordenada en y
y_num=f_num(x_num)

#    Mostrar gráfico
plt.plot(x_num,y_num)
plt.plot(c[0],f_num(c[0]),'*y',ms=15) # ms=marksize
plt.grid()
````

Ver diferencias entre f y f_num

````python
f(0) # 'f' es una variable simbólica
````

````python
f_num(0) # 'f_num' es una función numérica
````

````python
f_num(c[0])
````

**Resultado:**

En la función $f=(x-2)^2$, encontramos un mínimo global en $x=2$. Este mínimo es $f(2)=0$.

## 2. Criterio de la segunda derivada

Sea $f(x)$ una función tal que $f’(c)=0$ y cuya segunda derivada existe en un intervalo abierto que contiene a $c$.
- Si $f’’(c)>0$, entonces $f(c)$ es un mínimo relativo.
- Si $f’’(c)<0$, entonces $f(c)$ es un máximo relativo.
- Si $f’’(c)=0$, entonces el criterio no decide.

### Ejemplo

Mostrar, usando `sympy`, que la función $f(x)=(x-2)^2$ tiene un mínimo relativo en $x=2$.

Ya vimos que $f'(2)=0$. Notemos que:

````python
def f(x):
  return (x-2)**2
f(x)
````

````python
df=sym.diff(f(x),x)
c=sym.solve(df,x)
c
````

````python
# Segunda derivada
ddf=sym.diff(df,x)
ddf
````

````python
ddf=sym.diff(f(x),x,x)
ddf=sym.diff(f(x),x,2)
ddf
````

Por tanto, por el criterio de la segunda derivada, $f(2)=0$ es un mínimo relativo (en efecto, el mínimo global).

### Ejemplo
¿Qué pasa con $g(x)=(x-2)^3$ al intentar utilizar el criterio de la segunda derivada? (usar `sympy`).

````python
def g(x):
  return (x-2)**3
g(x)
````

````python
dg=sym.diff(g(x),x)
c=sym.solve(dg,x)
c
````

````python
ddg=sym.diff(dg,x)
ddg
````

````python
# Evaluamos 'ddg' en el punto critico x=c[0]
ddg.subs(x,c[0]) # c[0]=2
````

Como $g''(2)=0$ entonces el criterio de la segunda derivada no concluye.

````python
ddh=sym.diff(dh,x)
ddh
````

## 3. Método para determinar extremos absolutos de una función continua y=f(x) en [a,b]

- Determinar todos los valores críticos $c_1, c_2, c_3, \dots, c_n$ en $(a,b)$.
- Evaluar $f$ en todos los valores críticos y en los extremos $x=a$ y $x=b$.
- El más grande y el más pequeño de los valores de la lista $f(a), f(b), f(c_1), f(c_2), \dots, f(c_n)$ son el máximo absoluto y el mínimo absoluto, respectivamente, de $f$ en el intervalo $[a,b]$.

### Ejemplo

Determinar los extremos absolutos de $f(x)=x^2-2x+5$ en $\left[0,3\right]$.

Obtenemos los puntos críticos de $f$ en $\left[0,3\right]$:

````python
def f(x):
  return x**2-2*x+5
f(x)
````

````python
df=sym.diff(f(x),x)
c=sym.solve(df,x)
c
````

Evaluamos $f$ en los extremos y en los puntos críticos:

````python
x_num=np.array([0,1,3])
y_num=f(x_num)
y_num
````

Concluimos que el punto crítico $f(1)=4$ se trata de un mínimo en el intervalo $[0,3]$.

1. **Resistencia de una viga**  Una viga de madera tiene una sección transversal rectangular de altura $h$ y ancho $w$. La resistencia $S$ de la viga es directamente proporcional al ancho y al cuadrado de su altura ($S=kwh^2$, donde $k$ es una constante de proporcionalidad). ¿Cuáles son las dimensiones de la viga más fuerte que puede cortarse a partir de un leño redondo de $24$ pulgadas de diámetro?

````python
#@markdown Ayudín visual
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

C = patches.Circle((0, 0),12, fill=False,color='brown',lw=4)
S = patches.Rectangle((-9,-8),18,16,fill=False,color='k')
ax = plt.gca()
ax.add_artist(S)
ax.add_artist(C)

plt.text(0,8.5,'$w$')
plt.text(9.5,0,'$h$')

plt.plot([-9,9],[-8,8],'r--')
plt.text(0,2,'$24$')

plt.axis('equal')
plt.ylim([-15,15])
plt.axis('off')
plt.show()
````

**Solución:**

Se desea maximizar la resistencia $S=kwh^2$, sujeto a las dimesiones del tronco de diametro 24, esto es, aplicando el teorema de Pitágoras siendo 24 pulgadas la hipotenusa del triángulo rectángulo, se tiene que $24^2=w^2+h^2$.

Despejando $h$ y sustituyéndolo en $S$, la resistencia resulta
$$
\max_wS(w)=\max_wkw(24^2-w^2),
$$
para $0<w<24$.

````python
import sympy as sym
````

````python
sym.var('k')
sym.var('w',real=True)
def S(w):
  return k*w*(24**2-w**2)
S(w)
````

````python
dS=sym.diff(S(w),w)
dS
````

````python
c=sym.solve(dS,w)
c
````

````python
ddS=sym.diff(dS,w)
ddS
````

````python
ddS.subs(w,c[1])
````

````python
w_num=np.linspace(-15,15,500)
k=1
S_num=S(w_num)
````

````python
plt.plot(w_num,S_num)
for cr in c:
  plt.plot(cr,S(float(cr)),'*',ms=15)
plt.grid()
````

````python
# altura óptima de la viga 'h'
(24**2-c[1]**2)**0.5
````

````python
S(c[1])
````

**Resultado:**

Las dimensiones óptimas son $w=8\sqrt{3}$ y $h=19.60$ pulgadas, resultando en una resistencia máxima de $S=3072\sqrt{3}\ \
\rm{N\;in}$.

2. **Iluminación**  Una fuente luminosa se localiza sobre el centro de una mesa circular de $4 pies$ de diámetro. Encontrar la altura $h$ de la fuente luminosa de modo tal que la iluminación $I$ en el perímetro de la mesa sea máxima si $I=k(\sin \alpha)/s^2$, donde $s$ es la altura oblicua, $\alpha$ es el ángulo al cual la luz incide sobre la mesa y $k$ es una constante (El valor de esta última constante es irrelevante para el problema, ya que no se pide el valor de la iluminación).

````python
#@markdown Problema de iluminación
E = patches.Ellipse((0, 0),10, 3,fc='white',ec='brown')
C = patches.Circle((0,10.5),0.5,color='y')
a = patches.Arc((-5,0),2,2,theta2=60)
p1 = patches.Rectangle((-5, -2),0.5, 2, fill=False,color='brown')
p2 = patches.Rectangle((4.5, -2),0.5, 2, fill=False,color='brown')
p3 = patches.Rectangle((-0.25, -3),0.5, 2, fill=False,color='brown')
ax = plt.gca()
ax.add_artist(p1)
ax.add_artist(p2)
ax.add_artist(p3)
ax.add_artist(E)
ax.add_artist(C)
ax.add_artist(a)

plt.text(2,5,'$s$')
plt.text(0,3,'$h$')
plt.text(-4,0.5,r'$\alpha$')
plt.text(-0.5,-1,'$4$ pies')

plt.plot([-5,0,5],[0,10,0],'y--')
plt.plot([-5,5],[0,0],'k--')
plt.plot([0,0],[0,10],'k--')


plt.axis('equal')
plt.ylim([-3.1,12])
plt.axis('off')
plt.show()
````

**Solución:**

Queremos maximizar la iluminación $I=k\frac{\sin\alpha}{s^2}$.

Para establecer una relación entre $s$ y $h$, observamos un triángulo rectángulo entre el foco y el radio de la mesa, por lo que $s^2=h^2+2^2$, reemplazando $s^2$ en $I$, se tiene
$$
I=k\frac{\sin\alpha}{h^2+4}.
$$
Además, usando la definición de seno, sabemos que $\sin\alpha=h/s$ y que $s=\sqrt{h^2+4}$, sustituyendo resulta
$$
I=k\frac{h}{\sqrt{h^2+4}(h^2+4)},
$$
o de forma equivalente
$$
I=k\frac{h}{(h^2+4)^{3/2}},
$$
para $h\geq0$.

````python
import sympy as sym
````

````python
# Suponemos un valor para k=1
sym.var('h',positive=True)
def I(h):
  return h/(h**2+4)**1.5
I(h)
````

````python
dI=sym.diff(I(h),h)
c=sym.solve(dI,h)
c
````

````python
h_num=np.linspace(0,2,100)
I_num=I(h_num)
plt.plot(h_num,I_num)
plt.plot(c[0],I(c[0]),'*',ms=10)
plt.grid()
````

````python
# Iluminación máxima
I(c[0])
````

**Respuesta:**

Suponiendo una constante de proporcionalidad $k=1$, se tiene una iluminación máxima de $I=0.096$ a una altura de $h=1.4142$ pies.

3. **Fuerza mínima**  Se diseña un componente para deslizar un bloque de acero con peso $W=10 N$ a través de una mesa y sobre una canaleta. Se opone al movimiento del bloque una fuerza de fricción proporcional a su peso aparente (sea $k=0.1$ la constante de proporcionalidad). Determinar la fuerza mínima $f$ necesaria para deslizar el bloque y encontrar el valor correspondiente de $\theta$.

Ayudín: $f\cos \theta$ es la fuerza de dirección del movimiento, y $f\sin\theta$ es la fuerza que tiende a levantar el bloque. De tal modo, el peso aparente del bloque es $W-f\sin\theta$.

````python
#@markdown
W = patches.Rectangle((0, 2),0.8, 0.5, fill=False,color='k')
a = patches.Arrow(0.8,2.25,1,0.5,width=0.1,color='k')
arc = patches.Arc((0.8,2.25),1,1,theta2=30)
ax = plt.gca()
ax.add_artist(W)
ax.add_artist(a)
ax.add_artist(arc)

plt.text(0.2,2.6,'$W$')
plt.text(1.8,3,'$f$')
plt.text(1.35,2.305,r'$\theta$')

plt.plot([1.8,2.8],[2.75,3.25],'k')
plt.plot([0.8,2],[2.25,2.25],'k--')
plt.plot([-1,-1,2,2],[0,2,2,0],'k')


plt.axis('equal')
plt.ylim([0,4])
plt.axis('off')
plt.show()
````

**Solución:**

Para que el bloque se empiece a mover, la componente horizontal de $f$ debe igualar a la fricción, donde la fricción es $k(W-f\sin(\theta))$, esto es
$$
f\cos(\theta)=k(W-f\sin(\theta)),
$$
despejando se tiene el criterio de optimización a minimizar:
$$
\min_\theta f=\frac{kW}{\cos(\theta)+k\sin(\theta)},
$$
donde $0<\theta<\pi/2$.

````python
sym.var('theta')
````

````python
sym.var('k')
sym.var('W')
f=k*W/(sym.cos(theta)+k*sym.sin(theta))
f
````

````python
df=sym.diff(f,theta)
c=sym.solve(df,theta)
c
````

````python
float(c[0].subs(k,0.1))
````

````python
f_num=sym.lambdify(theta,f.subs({k:0.1,W:10}),'numpy')
th_num=np.linspace(0,0.2,100)
y_num=f_num(th_num)

plt.plot(th_num,y_num)
plt.plot(c[0].subs(k,0.1),f_num(float(c[0].subs(k,0.1))),'xr',ms=10)
plt.grid()
````

````python
f_num(float(c[0].subs(k,0.1)))
````

**Respuesta:**

Encontramos el ángulo óptimo en $\theta=0.09967$ radianes, con una fuerza mínima de $f=0.9950$ Newtons.

### En varias variables...

El procedimiento es análogo.

Si una función $f:\mathbb{R}^n\to\mathbb{R}$ alcanza un máximo o mínimo local en $\boldsymbol{x}=\boldsymbol{c}\in\mathbb{R}^n$, y $f$ es diferenciable en el punto $\boldsymbol{x}=\boldsymbol{c}$, entonces $\left.\frac{\partial f}{\partial \boldsymbol{x}}\right|_{\boldsymbol{x}=\boldsymbol{c}}=\boldsymbol{0}$ (todas las derivadas parciales en el punto $\boldsymbol{x}=\boldsymbol{c}$ son cero).

**Criterio de la segunda derivada:** para ver si es máximo o mínimo, se toma la segunda derivada (matriz Hessiana) y se verifica definición negativa o positiva, respectivamente.

Si se restringe a cierta región, hay ciertas técnicas. La más general, pero también la más compleja es la de **multiplicadores de Lagrange**.

**Ejemplo:** Determinar los valores extremo de $z(x,y)= (x-3)^2 + (y+5)^2$ usando el criterio de la segunda derivada y corroborarlo con la gráfica.

````python
sym.var('x')
sym.var('y')

def z(x,y):
  return (x-3)**2+(y+5)**2
z(x,y)
````

````python
dz_x=sym.diff(z(x,y),x)
dz_x
````

````python
dz_y=sym.diff(z(x,y),y)
dz_y
````

````python
c=sym.solve([dz_x,dz_y],[x,y])
c
````

````python
# Puntos críticos
c[x],c[y]
````

````python
# Segundas derivadas
#
# dz_x
dz_xx=sym.diff(dz_x,x)
dz_xy=sym.diff(dz_x,y) # d(dz/dx)/dy = d(dz/dy)/dx

# dz_y
dz_yy=sym.diff(dz_y,y)

H=sym.Matrix([[dz_xx,dz_xy],
              [dz_xy,dz_yy]])
H
````

````python
H.det()
````

````python
z(3,-5)
````

Encontramos que la función $z(x,y)=0$ es un mínimo evaluando en $x=3$ y $y=-5$.
